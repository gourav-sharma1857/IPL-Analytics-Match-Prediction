```python
def extract_match_info(filepath):
    with open(filepath, "r") as f:
        match = json.load(f)
    
    info = match["info"]
    match_id = os.path.basename(filepath).replace(".json", "")
    
    outcome = info.get("outcome", {})
    
    return {
        "match_id": match_id,
        "date": info["dates"][0],
        "city": info.get("city"),
        "venue": info.get("venue"),
        "team1": info["teams"][0],
        "team2": info["teams"][1],
        "toss_winner": info.get("toss", {}).get("winner"),
        "toss_decision": info.get("toss", {}).get("decision"),
        "winner": outcome.get("winner"),
        "result": outcome.get("result"),  # e.g. "tie", "no result" — None if normal win
    }
```

```python
def extract_deliveries(filepath):
    with open(filepath, "r") as f:
        match = json.load(f)
    
    match_id = os.path.basename(filepath).replace(".json", "")
    rows = []
    
    for inning_num, inning in enumerate(match["innings"], start=1):
        team = inning["team"]
        for over_data in inning["overs"]:
            over_num = over_data["over"]
            for ball in over_data["deliveries"]:
                rows.append({
                    "match_id": match_id,
                    "innings": inning_num,
                    "innings_team": team,
                    "over": over_num,
                    "ball": ball["actual_delivery"],
                    "batter": ball["batter"],
                    "bowler": ball["bowler"],
                    "non_striker": ball["non_striker"],
                    "runs_batter": ball["runs"]["batter"],
                    "runs_extras": ball["runs"]["extras"],
                    "runs_total": ball["runs"]["total"],
                    "is_wicket": "wickets" in ball,
                })
    
    return rows
```

```python
sample_path = os.path.join(raw_dir, sample_file)

info_test = extract_match_info(sample_path)
deliveries_test = extract_deliveries(sample_path)

print(info_test)
print(len(deliveries_test), "deliveries")
```

```python
all_files = [f for f in os.listdir(raw_dir) if f.endswith(".json")]
print(len(all_files), "match files found")
```

```python
all_matches = []
all_deliveries = []
failed_files = []

for filename in all_files:
    filepath = os.path.join(raw_dir, filename)
    try:
        match_info = extract_match_info(filepath)
        match_deliveries = extract_deliveries(filepath)
        
        all_matches.append(match_info)
        all_deliveries.extend(match_deliveries)
    except Exception as e:
        failed_files.append((filename, str(e)))

print(f"Successfully processed: {len(all_matches)} matches")
print(f"Failed: {len(failed_files)} files")
if failed_files:
    print(failed_files[:5])  # show first few failures
```

```python
matches_df = pd.DataFrame(all_matches)
deliveries_df = pd.DataFrame(all_deliveries)

print(matches_df.shape)
print(deliveries_df.shape)

matches_df.head()
```
